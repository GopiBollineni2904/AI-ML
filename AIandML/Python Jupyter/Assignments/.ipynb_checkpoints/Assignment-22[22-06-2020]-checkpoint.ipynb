#### Question 2:

Mean = 70 and Standard Deviation = 10.
Identify how much area can be found between 70 and 90 with help of Empirical Rule.

```The value '90' is 20 units away from the mean '70', i.e., 2 S.D. (Standard deviations) away from the mean.
Therefore, according to the EMPIRICAL RULE, including 2 S.D. away from mean on either side of the mean, there is 95% area in total.
But, the area between 70 and 90 is on the right half of this. So, the required area is 95/2 ,i.e., 47.5% of the total area.```
