```python
#Apply the KNN Algorithm to the dataset without applying scaling operation
#and try to compare the results before scaling and after scaling.
X=df.drop(['diagnosis'],axis=1)
y=df.diagnosis

#splitting the data:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

### Before SCALING:

```python
#first let us find the optimum K-value using ELBOW method:
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
score=[]
for k in range(1,20):
    knn_model=KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train,y_train)
    pred_test=knn_model.predict(X_test)
    score.append(accuracy_score(y_test,pred_test))
```

```python
#plotting the score:
plt.plot(range(1,20),score,marker='^')
```

```python
#optimum value for K is 11
final_model=KNeighborsClassifier(n_neighbors=11,metric='euclidean')
final_model.fit(X_train,y_train)
```

```python
y_train_pred=final_model.predict(X_train)
y_test_pred=final_model.predict(X_test)
#lets keep it like this and compare both the calssification reports in the end
#just for more detail between before and after scaling
```

## After SCALING:

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
scaled_X_train=pd.DataFrame(data=ss.fit_transform(X_train),columns=X_train.columns)
scaled_X_test=pd.DataFrame(data=ss.fit_transform(X_test),columns=X_test.columns)
```

```python
#first let us find the optimum K-value using ELBOW method:
score1=[]
for k in range(1,20):
    scaled_knn_model=KNeighborsClassifier(n_neighbors=k)
    scaled_knn_model.fit(scaled_X_train,y_train)
    scaled_pred_test=scaled_knn_model.predict(scaled_X_test)
    score1.append(accuracy_score(y_test,scaled_pred_test))
```

```python
#plotting the score:
plt.plot(range(1,20),score1,marker='o')
```

```python
#optimum value for K is 8
scaled_final_model=KNeighborsClassifier(n_neighbors=8,metric='euclidean')
scaled_final_model.fit(scaled_X_train,y_train)
```

```python
scaled_y_train_pred=scaled_final_model.predict(scaled_X_train)
scaled_y_test_pred=scaled_final_model.predict(scaled_X_test)
```

## Classification Reports for Train & Test data:
#### Before scaling:

```python
print('CLASSIFICATION REPORT before SCALING:\n')
print('on TRAINING DATA:\n')
print(classification_report(y_train_pred,y_train))
print('\non TESTING DATA:\n')
print(classification_report(y_test_pred,y_test))
print('\nCLASSIFICATION REPORT after SCALING:\n')
print('on TRAINING DATA:\n')
print(classification_report(scaled_y_train_pred,y_train))
print('\non TESTING DATA:\n')
print(classification_report(scaled_y_test_pred,y_test))
```

## Observations on performing applying the KNN algo before & after scaling the data:
`
Before scaling--> the Accuracy of the model was 94%
After scaling --> the accuracy of the model was 97%
Now, that might not look like much of an increase in the accuracy of the model, some might even think that it doesn't make much difference but, slightest change in accuracy of machine learning models is considered quite significant.
That being said, here, we can say that the model that was trained on scaled values worked much better than the one that was trained on non-uniform values.
This indicates that, distance based algorithms like KNN produce better outcomes when provide with scalled data, rather than test data.
