## Renaming Columns:

```python
import pandas as pd
import numpy as np
s1=pd.Series([9,1,8,2,7,3,4,6,5,10])
s2=pd.Series([1.1,3.5,4.7,5.8,2.9,9.3,5.4,7.5,3.9,4.5])
s3=pd.Series(['a','b','c','d','e','f','g','h','i','j'])
Data= {'first':s1,'second':s2,'third':s3}#define data
df=pd.DataFrame(Data)#creating a DataFrame
df
```

```python
df.rename(columns={'second':'column2','third':'column3'},inplace=True)
```

```python
df #the given old column/row names doesn't match, rename function doesn't do anything
```

## Changing the row index:

### Removing single or Multiple Columns:

```python
import pandas as pd
import numpy as np
s1=pd.Series([9,1,8,2,7,3,4,6,5,10])
s2=pd.Series([1.1,3.5,4.7,5.8,2.9,9.3,5.4,7.5,3.9,4.5])
s3=pd.Series(['a','b','c','d','e','f','g','h','i','j'])
Data= {'first':s1,'second':s2,'third':s3}#define data
df=pd.DataFrame(Data)#creating a DataFrame
df
```

```python
#single column removing
#df.drop(['first'],axis=1)
df.drop(columns=['first'])
```

```python
df.drop(columns=['first','second'])
```

```python
df.drop(df.columns[[0,2]],axis=1)
```

### removing Rows:

```python
#single Row
df.drop([7])
```

```python
df.drop([7,8])
```

## Working With CSV files:

```python
df1=pd.read_csv('diamonds.csv',index_col=0)
df1 #index denotes the position of the INDEX values
```

```python
df1=pd.read_csv('diamonds.csv',index_col=0,nrows=10,na_values=" ")
df1 #nrows--> no. of rows to include; na_value -->Not available Values(whe the string in the quotes is found it is considered to be a Null Value by Pandas, this is done as PANDAS is not Capable of detecting NULL values in case of any custom values in the place of Null)
```

```python
df1.shape #returns shae of the CSV table
```

```python
pd.isnull(df1) #returns True for a null value and False for a notNull value in the CSV table
```
