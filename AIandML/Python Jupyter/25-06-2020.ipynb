```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading Boston dataset
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
df=pd.DataFrame(boston.data,columns=boston.feature_names)
df.head(10)
```

```python
df['MEDV']=boston.target
df.head(10)
```

```python
## to find the correlation we use a function called corr()
df.corr()
```

```python
plt.subplots(figsize=(14,14))
sns.heatmap(df.corr(),annot=True,cmap='YlGnBu')
```

```python
plt.subplots(figsize=(14,14))
sns.heatmap(df.corr()[df.corr()>=0.7],annot=True,cmap='YlGnBu')
```

```python
#all the columns as input
# DRop the columns which are highly corelated and build a MLR model u need to compare the performance
```

```python
# splitting the data into train and test
X=df.drop(['MEDV'],axis=1)
y=df.MEDV
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
#Intercept and coeff values
print(lr.intercept_)
print(lr.coef_)
```

```python
#create a dataframe for coefficients
coefficients = pd.DataFrame([X_train.columns,lr.coef_]).T
coefficients
```

```python
#checking the predictions on training data
y_train_pred=lr.predict(X_train)
y_train_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error 
print('R^2:', r2_score(y_train,y_train_pred)) 
print('Adjusted R^2:', 1- (1-r2_score(y_train, y_train_pred))*(len(X_train)-1)/
      (len(X_train)-X_train.shape[1]-1)) 
print('MAE: ',mean_absolute_error(y_train,y_train_pred))
print('MSE: ',mean_squared_error(y_train,y_train_pred))
print('RMSE: ',np.sqrt(mean_squared_error(y_train,y_train_pred)))
```

```python
#plotting
sns.regplot(y_train,y_train_pred)
```

```python
#visualize the differences between actual and predicted values
plt.scatter(y_train,y_train_pred)
plt.xlabel('Price')
plt.ylabel('Predicted Price')
plt.title('Diff in Actual n Predicted')
```

```python
#plotting between y_train and error
plt.scatter(y_train,y_train-y_train_pred)
plt.xlabel('Price')
plt.ylabel('Predicted Price')
plt.title('Diff in Actual n Errors[Residuals]')
```

```python
#testing the model on test data
y_test_pred=lr.predict(X_test)
y_test_pred
```

```python
from sklearn.metrics import r2_score ,mean_absolute_error,mean_squared_error
print('R^2:', r2_score(y_test,y_test_pred)) 
print('Adjusted R^2:', 1- (1-r2_score(y_test, y_test_pred))*(len(X_test)-1)/
      (len(X_test)-X_test.shape[1]-1)) 
print('MAE: ',mean_absolute_error(y_test,y_test_pred))
print('MSE: ',mean_squared_error(y_test,y_test_pred))
print('RMSE: ',np.sqrt(mean_squared_error(y_test,y_test_pred)))
```
