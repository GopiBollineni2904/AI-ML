### OneHotEncoder:
###### similar to--> pd.get_dummies()

- Import OneHotEncoder from sklearn
- Import ColumnTransformer from sklearn
- create object for ColumnTransformer to pass the OneHotEncoder into this
- ColumnTransformer has 2 mandatory args--> transformers,remainder
<br>
--> transformers arg accepts list of tuples

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
column_transformer=ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[0])],
                                     remainder='passthrough')
```

```python
data = column_transformer.fit_transform(data)
data
```

```python
X=pd.DataFrame(data)
X
```

```python
# Importing the Classes
## Creating the objects
## To apply the object to the dataframe(columns), we are using fit_transform--> With help of fit_transform
## we are encoding the categorical variables
## LabelEncoder is used only when we have to classes
```
