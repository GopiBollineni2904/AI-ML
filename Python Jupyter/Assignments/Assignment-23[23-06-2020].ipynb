```python
#dataset --> https://www.kaggle.com/benroshan/factors-affecting-campus-placement/data
#Campus Placement Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df=pd.read_csv('Placement_Data_Full_Class.csv',index_col='sl_no')
df
```

```python
#Check the data types and get the summary stats
df.dtypes
```

```python
# summary stats:
df.describe(include='all')
```

```python
# If you have any missing values in the numerical columns try imputing those columns     
    #-Create a dataframe by imputing  the missing values with mean 
    #-Create a dataframe by imputing the missing values with median
df.isnull().sum() #only salary column has '67' missing values the rest are fine
```

```python
#imputing the salary column:
    #-using mean:
df1=df.salary.fillna(df.salary.mean())
df1.head(10)
```

```python
    #-using median:
df.salary.fillna(df.salary.median(),inplace=True)
df.head(10)
```

```python
#If any missing values in categorical columns, impute it with mode
df.isna().sum() #no missing Values
```

```python
#For the categorical columns apply  different kinds of encoding techniques
df.select_dtypes(include='object')
```

```python
#gender --> LabelEncoder
#ssc_b --> LabelEncoder
#hsc_b --> LabelEncoder
#hsc_s --> get_dummies
#degree_t --> OneHotEncoder
#workex --> LabelEncoder
#specialisation --> LabelEncoder
#status --> LabelEncoder
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
le= LabelEncoder()
oh= OneHotEncoder()
```

```python
en_df=df.copy()
#LabelEncoder
en_df.gender=le.fit_transform(df.gender) #gender
en_df.ssc_b=le.fit_transform(df.ssc_b) #ssc_b
en_df.hsc_b=le.fit_transform(df.hsc_b) #hsc_b
en_df.workex=le.fit_transform(df.workex) #workex
en_df.specialisation=le.fit_transform(df.specialisation) #specialisation
en_df.status=le.fit_transform(df.status) #status
#get_dummies
dummy1=pd.get_dummies(en_df.hsc_s,drop_first=True,prefix='hsc_s')
dummy2=pd.get_dummies(en_df.degree_t,drop_first=True,prefix='degree_t')
en_df=pd.concat([en_df,dummy1,dummy2],axis=1) #hsc_s
```
