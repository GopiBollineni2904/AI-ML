````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
%matplotlib inline
data=pd.read_csv(r'https://raw.githubusercontent.com/kotagiri-preeti/Summer-Internship-AI-and-ML/master/Match%20(1).csv')
````

````python
df=pd.DataFrame(data)
df.head(10)
````

```Question 1:

```

````python
g=df.groupby('ManOfMach')
g.describe()
````

````python
#Answer ====>>
g.Match_SK.count()
````

```Question 2:
Plot a pie chart for the country_name column
```

````python
h=df.groupby('Country_Name')
h.describe()
````

````python
##Answer
plot.pie(h.Match_SK.count(),labels=['India','South Africa','U.A.E'],data='data',autopct='%0.1f%%',
         explode=[0.2,0,0],radius=2,startangle=45,colors=('b','g','r'))
plot.show()
````

````python
##Alternate
plot.pie(df['Country_Name'].value_counts(),labels=['India','South Africa','U.A.E'],data='data',autopct='%0.1f%%',
         explode=[0.2,0,0],radius=2,startangle=45,colors=('b','g','r'),pctdistance=0.8)
plot.grid()
````

````python
list(h.groups.keys())
````

````python
## Another Alternative to pass the labels
plot.pie(df['Country_Name'].value_counts(),labels=list(h.groups.keys()),data='data',autopct='%0.1f%%',
         explode=[0.2,0,0],radius=2,startangle=45,colors=('b','g','r'),pctdistance=0.8)
plot.grid()
````
