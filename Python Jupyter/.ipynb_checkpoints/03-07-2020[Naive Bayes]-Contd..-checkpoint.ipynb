```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#reading data
df=pd.read_csv('https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/spam.csv',encoding='latin-1')
#there are some emojis and some special symbols in the data in YTF-8 format
#so, change the text encoding in the data set to 'latin-1'
df.head()
```

```python
df.shape
```

```python
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
```

```python
#split the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.v2,df.v1,test_size=0.2,random_state=1)
```

```python
#CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
count_vect=CountVectorizer()
```

```python
word_count_vector=count_vect.fit(X_train)
#to get feature names
word_count_vector.get_feature_names()
```

```python
doc_array=word_count_vector.transform(X_train).toarray()
doc_array
```

```python
#converting the array into a dataframe
feature_matrix=pd.DataFrame(data=doc_array,columns=word_count_vector.get_feature_names())
feature_matrix
```

# 03-07-2020: [Naive Bayes Contd...]

```python
#TFIDF Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
```

```python
#fitting the model
X_train_trans=tfidf.fit_transform(X_train)
X_train_trans
```

```python
X_test_trans=tfidf.transform(X_test)
X_test_trans
```

```python
#Feature names
tfidf.get_feature_names()
```

```python
#position of words
tfidf.vocabulary_
```

```python
#idf of the terms
tfidf.idf_ #Inverse Document Frequency
```

```python
#Import BernoulliNB
from sklearn.naive_bayes import BernoulliNB
model_BernNB=BernoulliNB()
```

```python
#applying the model
model_BernNB.fit(X_train_trans,y_train)
```

```python
#Prediction on test data
y_test_pred=model_BernNB.predict(X_test_trans)
```

```python
#comapre the actual and predicted values
from sklearn.metrics import classification_report, confusion_matrix
confusion_matrix(y_test,y_test_pred)
```

```python
print(classification_report(y_test,y_test_pred))
```

```python
#test a single message
new_message=pd.Series('Last few days to collect the free\
                        entry passes with 40% discount')
#Calculate TFIDF values for the new message
new_message_trans=tfidf.transform(new_message)
```

```python
#predict the message
model_BernNB.predict(new_message_trans)
```
