# Merging : inner,outer,left,right:

```python
df1 = pd.DataFrame({"branch":['hyd','blr','vizag','pune'], #like a dictionary
                   "total":[90,96,98,92]})
df1
```

```python
df2 = pd.DataFrame({"branch":['blr','hyd','vizag','chennai'],
                   "distance_from_city":[40,60,12,20]})
df2
```

```python
pd.merge(df1,df2,on="branch",how='inner') ##returns the columns that are available in both DataFrames with the data
```

```python
pd.merge(df1,df2,on="branch") #default is inner
```

```python
pd.merge(df1,df2,on="branch",how='outer') #merges with all the row fiels but the data unavailable as NaN
```

```python
pd.merge(df1,df2,on="branch",how='right') #df1 is left and df2 is right[right merge- df2 merger happens; left merge- df1]
```

```python
pd.merge(df1,df2,on="branch",how='left',indicator=True)
#'idicator' attribute indicate the presence of the Columns in both tables
```

## groupby:

```python
data = {"name":['a','b','c','d','e','f'],
       "branch":['hyd','vizag','blr','hyd','vizag','blr'],
       "marks":[90,91,92,93,94,95]}
df = pd.DataFrame(data)
df
```

```python
g = df.groupby("branch")
g
```

```python
for branch,branch_df in g:
    print(branch)
    print(branch_df)
```

```python
g.get_group('hyd')["marks"].sum()
```

```python
g.get_group('hyd')["marks"].mean()
```

```python
g.mean()
```
