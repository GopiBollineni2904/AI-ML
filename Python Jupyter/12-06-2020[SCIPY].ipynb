### Linear Algebra:

```python
from scipy import linalg
#create 2d matrix
ar2=np.array([[4,5],[3,2]])
linalg.det(ar2) #determinant of the matrix
```

```python
#inverse of the matrix
linalg.inv(ar2)
```

## Image Operations:

```python
from scipy import misc
data=misc.face()
plt.imshow(data)
```

```python
data.shape,type(data)
```

```python
#to flip the image
flip_data= np.flipud(data)
plt.imshow(flip_data)
```

```python
rand_arr = np.random.randint(size=data.shape,low=0,high=255)
plt.imshow(rand_arr)
```

```python
##to Rotate the image
from scipy import ndimage
data_rotate = ndimage.rotate(data,angle=135)
plt.imshow(data_rotate)
```

```python
data_rotate = ndimage.rotate(data,angle=45)
plt.imshow(data_rotate)
```
